```python
from fer import FER
import cv2

img = cv2.imread("img1.jpeg")
detector = FER()
result = detector.detect_emotions(img)
print(result)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 이미지 출력
plt.imshow(img_rgb)
plt.axis('off')  # 축 제거
plt.show()
```
